```python
import pandas as pd
```

```python
df = pd.read_csv('News_York_data.csv')
```

This csv contain the 1660 alticles of reporting about sexual harassment on NYT. (when searching the key word "sexual harassment"on NYT websit, the reuslt show there are 1660)

```python
df 
```

```python
df['title'][1].split()
```

```python
all_title=[]
for title in df['title']:
    all_title.extend(str(title).split())
```

```python
all_title
```

```python
word_count={}
```

```python
for word in all_title:
    if word not in word_count:
        word_count[word]=1
    else:
        word_count[word]=word_count[word]+1
```

```python
word_count
```

```python
df_word_count=pd.Series(word_count).to_frame().reset_index()
```

```python
df_word_count.sort_values(by=0,ascending=False)
```

```python
stop_words=set([
    'of',
    'the',
    'a',
    'and',
    'in',
    'your',
    'for',
    'is',
    'on',
    'A',
    'With',
    'at',
    'What',
    'How',
    'about',
    'Dealbook',
    'Is',
    'Your',
    'New',
    'to',
    'The',
    'After',
    'Today:',
    'by',
    'it',
    'as',
    'Not',
    'Are',
    'Who',
    'Briefing',
    'In',
    'York',
    'over',
    'From',
    'It',
    'About',
    'Over',
    'Will',
    'Tuesday',
    'Monday',
    'Wednesday',
    'Thursday',
    'We',
    '2017',
    'DealBook',
    'his',
    'Be',
    '‘The',
    'Times',
    'an',
    'This',
    'Sexual',
    'Harassment',
    'Donald',
    'Harvy'
     
])
```

removing some preposition as well as some high frequency words without meaningless,like the name of the papar"New""York""Times".

```python
def is_not_stop_word(x):
    return x not in stop_words
df_clean_word=df_word_count[
    df_word_count['index'].apply(is_not_stop_word)
].sort_values(by=0,ascending=False)
```

```python
df_clean_word
```

```python
temp_dict = {}
for index, value in  df_clean_word.as_matrix():
    temp_dict[index] = value
```

```python
temp_dict
```

```python
from wordcloud import WordCloud
wc=WordCloud
```

```python
w=wc().generate_from_frequencies(temp_dict)
```

```python
import matplotlib.pyplot as plt
plt.imshow(w)
plt.axis('off')
```

```python
import pandas as pd
import numpy as np
import json
import jieba
from matplotlib import pyplot as plt
from pandas import Series,DataFrame
```

```python
with open('cooking_xiang.json', 'r') as f:
    data1 = json.load(f)
df = pd.DataFrame(data1)
```

```python
df
```

```python
all_ingredient=df['recipe'].value_counts().index
all_ingredient
```

```python
all_ingredient=[]
for recipe in df['recipe']:
    x=list(jieba.cut(recipe))
    all_ingredient.extend(x)
```

```python
len(all_ingredient)
```

```python
igredient = list(set(all_ingredient))
len(igredient)
```

```python
def get_material_value1(v):
    temp_data = []
    for r in igredient:
        if r in v:
            temp_data.append(1)
        else:
                temp_data.append(0)
    return temp_data
```

```python
df['material-value'] = df['recipe'].apply(get_material_value1)
df['material-value']
```
